```python
def bar_plot(dataset, feature, stacked= True):
    survived = dataset[dataset["Survived"] == 1][feature].value_counts()
    dead = dataset[dataset["Survived"] == 0][feature].value_counts()
    df_survived_dead = pd.DataFrame([survived, dead])
    df_survived_dead.index = ['survived', 'dead']
    pt= df_survived_dead.plot(kind = "bar", stacked= stacked)
```

```python
def plot_barplot(dataset, feature, title, categories):
    fig, ax= plt.subplots(figsize =(5,5))
    fare_range = pd.qcut(dataset[feature], len(categories), labels=categories)
    ax.set_title(title)
    sns.barplot(x=fare_range, y=dataset.Survived, ci=None).set_ylabel('Survival Rate')

```

```python
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestClassifier


num_col = ['Fare']
cat_col = ['Pclass','Title', 'Embarked', 'Fam_type', 'Ticket_length', 'Ticket_2Letters_start']

num_transformer = SimpleImputer(strategy = 'median')
cat_transformer = Pipeline(steps= [
    ('imputer',SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', num_transformer, num_col),
        ('cat', cat_transformer, cat_col)
    ]
)

titanic_pipline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', RandomForestClassifier(random_state=0, n_estimators=500, max_depth=5))    
])

#Training
titanic_pipline.fit(x_train, y_train)
```

```python
from sklearn.model_selection import cross_val_score
print('Cross validation score', cross_val_score(titanic_pipline, x_train, y_train, cv=10).mean())
```
